```python
import pandas as pd
import numpy as np
import cv2
import os
CWD = os.getcwd()
print(CWD)
```

metadata_ISIC_ArGI already contains the label, malignance, and skin_tone categories. all  i have to do is fix the indexes, ids, content, and style.

```python
### Remove Vanilla ISIC Images ###
df = pd.read_excel(f'{CWD}/metadata_ISIC_ArGI.xlsx', index_col='index')
print(df)
badids = []
for id in df['id']:
    if (id.find("stylized_with") == -1):
        badids.append(id)
print(badids)
print(len(badids))
```

```python
df = pd.read_excel(f'{CWD}/metadata_ISIC_ArGI.xlsx', index_col='id')
# print(df)
df.drop(labels = badids, inplace=True)
df.drop(['index'], axis = 1, inplace=True)

# print(df)
# print(len(df['label']))

n_ArGI = len(df['label'])
print(n_ArGI)

### Create Metadata for ArGI ###
indexes = np.arange(0,n_ArGI)
images= np.empty(shape=[n_ArGI, 64, 64, 3], dtype=np.uint8)
id_nums = np.full(shape=[n_ArGI,1], fill_value=100000)
ids = np.empty(shape=[n_ArGI], dtype=np.dtypes.StrDType)
labels = np.empty(shape=[n_ArGI], dtype=np.dtypes.StrDType)
malignances = np.empty(shape=[n_ArGI])
skin_tones = np.empty(shape=[n_ArGI])
contents = np.empty(shape=[n_ArGI], dtype=np.dtypes.StrDType)
styles = np.empty(shape=[n_ArGI], dtype=np.dtypes.StrDType)

#print(indexes)
df = df.assign(index=indexes)
df['id'] = df.index
df.set_index('index', inplace=True)
print(df)
```

```python
### create filenames ###
"""
Start with 100 000
add the index of the image
then change leading digit to 0
now you have a six-digit id.
"""
```

```python
for i in range(1, n_ArGI+1):
    idx = i-1
    old_name = df['id'][idx]
    labels[idx] = df['label'][idx]
    malignances[idx] = df['malignance'][idx]
    skin_tones[idx] = df['skin_tone'][idx]
    contents[idx] = old_name[0:11] + '.jpg'
    styles[idx] = old_name[29:-4] + '.jpg'
    id_nums[idx] = id_nums[idx] + i

    id_str = str(id_nums[idx])
    id_str = f'0{id_str[2:-1]}'
    print(id_str)
    new_name = f'ArGI_{id_str}.jpg'
    print(new_name)

    ids[idx] = new_name
    os.rename(f'{CWD}/all_images/{old_name}', f'{CWD}/all_images/{new_name}.jpg')
```

```python
print(indexes.shape)
print(ids.shape)
print(labels.shape)
print(malignances.shape)
print(skin_tones.shape)
print(contents.shape)
print(styles.shape)
```

```python
metadata_ArGI = {
    'index': indexes,
    'id': ids,
    'label': labels,
    'malignance': malignances,
    'skin_tone': skin_tones,
    'content': contents,
    'style': styles
}
metadata_ArGI = pd.DataFrame(metadata_ArGI)
metadata_ArGI.set_index('index')
metadata_ArGI.drop('index', axis=1, inplace=True)
print(metadata_ArGI)
```

```python
metadata_ArGI.to_excel(f'{CWD}/metadata_ArGI.xlsx')
```
