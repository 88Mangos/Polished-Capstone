```python
from PIL import Image
from io import BytesIO
import pandas as pd
from bs4 import BeautifulSoup
import os
import numpy as np
import requests
```

```python
CWD = os.getcwd()
```

```python
urls = pd.read_excel(f'{CWD}/urls_to_scrape.xlsx')
print(urls)
```

```python
malignant_urls = urls['malignant']
malignant_urls = malignant_urls.to_numpy()
malignant_urls = malignant_urls[0:15]
print(malignant_urls)
print(len(malignant_urls))

benign_urls = urls['benign']
benign_urls = benign_urls.to_numpy()
print(benign_urls)
print(len(benign_urls))
```

```python
def download_image(url, folder_path, image_name):
    response = requests.get(url)
    if response.status_code == 200:
        img = Image.open(BytesIO(response.content))
        img.save(f'{folder_path}/{image_name}.jpg')
        print(f'Image {image_name} downloaded successfully.')
    else:
        print(f'Failed to download image {image_name}. Status code: {response.status_code}')
```

```python
def scrape_images(url, label, folder_path):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        images = soup.find_all('img')
        for img in images:
            img_url = img.get('src')
            img_url = img_url[0:3] + img_url[img_url.find('imageId')-1:]
            imgId = int(img_url[img_url.find('imageId')+len('imageId')+1:-8])
            imgId += 100000 # as of 5/30/2024 the atlas has 12,351 pictures.
            # a 6-digit id is more than sufficient, allowing for consistent format.
            imgId = str(imgId)
            image_name = f'BrAD_0{imgId[1:]}'
            img_url = 'https://www.atlasdermatologico.com.br/' + img_url
            print(imgId)
            print(img_url)
            print(image_name)
            if (img_url):
                download_image(img_url, folder_path, image_name)
```

```python
for url in benign_urls:
    scrape_images(url, 'benign', f'{CWD}/benign')
```

```python
for url in malignant_urls:
    scrape_images(url, 'malignant', f'{CWD}/malignant')
```
